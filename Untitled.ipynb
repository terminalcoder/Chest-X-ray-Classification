## Overview
This report aims to construct an image classification model capable of accurately diagnosing pneumonia in pediatric patients, from chest X-ray images. Additionaly, such a tool would enable medical professionals to delegate such tasks to automated systems to attain greater efficiency in healthcare.

## Data  
The images used for this project come from the Kaggle 

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from keras.preprocessing.image import ImageDataGenerator
from keras import layers, models
```

```python
train_dir = 'data/chest_xray/train'
validation_dir = 'data/chest_xray/val'
test_dir = 'data/chest_xray/test'
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(train_dir,
                                                    target_size=(150, 150),
                                                    batch_size=20,
                                                    class_mode='binary')

val_generator = val_datagen.flow_from_directory(validation_dir,
                                                target_size=(150, 150),
                                                batch_size=20,
                                                class_mode='binary')
```

View label mapping:

```python
train_generator.class_indices
```

Visualize distribution of training data:

```python
# create array of classes
tr_dist_counts = np.bincount(train_generator.classes)


# visualize distribution of training data
plt.bar(train_generator.class_indices.values(), tr_dist_counts, tick_label = [x.capitalize() for x in train_generator.class_indices.keys()], color = ['lightsteelblue', 'cornflowerblue'])
plt.ylabel = 'count'
plt.title('Distribution of Training Data Images');
```

To get a sense of what we are training the model to do, display a few examples of a normal scan & examples of pnuemonia scan:

```python
img, label = train_generator[1]

fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15, 10))

x = 0
y = 0

for i in range(20):

    if label[i] == 0 and x < 4:
        axes[0, x].imshow(img[i])
        axes[0, x].set_title('normal')
        x += 1
        
    if label[i] == 1 and y < 4:
        axes[1, y].imshow(img[i])
        axes[1, y].set_title('pnuemonia')
        y += 1
```
